```python
# global variable = outside the func(
x = "chethan"
def myfunc():
    global x
    x = "banana"
myfunc()
print("hello.." +x)
```

```python
# local variabl = inside the function
x = "chethan"
def myfunc():
    x = "banana"
myfunc()
print("hello.." +x)
```

```python
class bird:
    def __init__(self):
        self.name = "gubbi"
        self.color = "black"
        self.sound = "chew chew"
        self.age = 2
    def fly(self):
        print("bird will fly and search for food")
b1 = bird()
print(b1.name)
print(b1.color)
print(b1.sound)
print(b1.age)
b1.fly()
# print(b1)
```
